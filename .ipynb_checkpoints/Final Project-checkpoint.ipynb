## II. Prepare modeling database

### Merge relevent datasets together

```python
j.head()
```

```python
# create genre dummies
md = j.join(pd.get_dummies(j['genre'], prefix='genre'))
```

```python
# merge review text to modeling DB
md = md.merge(c, how='left', on='reviewid')
```


### Use count vectorizer to append most common ngrams to modeling dataframe

```python
reviews = md['content'].fillna('')

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features = 1000, 
                             ngram_range=(1, 3), 
                             stop_words='english',
                             binary=False)

vectorizer.fit(reviews)
X = vectorizer.transform(reviews).toarray()
```

```python
tokenized = pd.DataFrame(data=X, columns=vectorizer.get_feature_names())
tokenized.sum().sort_values(ascending=False)
```

```python
md1 = md.join(tokenized, rsuffix='txt')
```

```python
md1.head()
```

```python
md1.info()
```

### Also create a second version of the modeling dataset that uses Tfidf vectorizer, called `m2`

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features = 1000, 
                             ngram_range=(1, 3), 
                             stop_words='english',
                             binary=False)

vectorizer.fit(reviews)

X = vectorizer.transform(reviews).toarray()
```

```python
tokenized = pd.DataFrame(data=X, columns=vectorizer.get_feature_names())
tokenized.sum().sort_values(ascending=False)
```

```python
md2 = md.join(tokenized, rsuffix='txt')
```

```python
md2.head()
```

```python
md2.info()
```

### III. Modeling

Now we have two modeling datasets, `md1` and `md2` which use countvectorizer and Tfidfvectorizer, respectively. Can we build a logistic regression based on each?

```python
data = md1
```

```python
features = data.iloc[:, 14:]
del features['content']
features.head()
```

```python
label = data['best_new_music']
```

```python
from sklearn import grid_search, cross_validation
from sklearn.linear_model import LogisticRegression

gs1 = grid_search.GridSearchCV(
    estimator=LogisticRegression(),
    param_grid={'C': [10**-i for i in range(-5, 5)], 'class_weight': [None, 'auto']},
    cv=cross_validation.KFold(n=len(features), n_folds=3),
    scoring='roc_auc'
)

gs1.fit(features.as_matrix(), label)
```

```python
gs1.best_score_
```

### Can a regression using Tfidf vectorizer perform better?

```python
data = md2
```

```python
features = data.iloc[:, 14:]
del features['content']
features.head()
```

```python
label = data['best_new_music']
```

```python
from sklearn import grid_search, cross_validation

gs2 = grid_search.GridSearchCV(
    estimator=LogisticRegression(),
    param_grid={'C': [10**-i for i in range(-5, 5)], 'class_weight': [None, 'auto']},
    cv=cross_validation.KFold(n=len(features), n_folds=3),
    scoring='roc_auc'
)

gs2.fit(features.as_matrix(), label)
```

```python
gs2.best_score_
```

The AUC score is slightly improved by using Tfidf. 

### Which features have the biggest impact on predicted probability? 

Negative words like "unfortunately", "fine", "worse", etc. are associated with lower probability of an album being labeled "Best New Music." Words associated with old albums being re-released (like "reissue", "record", "vinyl", "box") were most-highly associated with probability of an album being labeled "Best New Music."

```python
coeffs2 = []
for i in gs2.best_estimator_.coef_.tolist():
    coeffs2.append(np.exp(i))
```

```python
pd.DataFrame(zip(features.columns.tolist(), coeffs2[0])).sort(columns=1, ascending=False)
```

### How well is this model classifying?

```python
yhat = gs2.best_estimator_.predict(features)
```

```python
yhat
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, precision_score, auc
confusion = confusion_matrix(y_true=md2['best_new_music'], y_pred=yhat)
print confusion
```

```python
accuracy = (confusion[0,0] + confusion[1,1]) / (confusion.sum() * 1.0)
print accuracy
```

```python
import sklearn.metrics as metrics

precision = metrics.precision_score(md2['best_new_music'], yhat) 
recall = metrics.recall_score(md2['best_new_music'], yhat)
print precision, recall
```

The current model yields a lot of false negatives -- we're classifying many true ones as 0. And the overall model accuracy is 95%. Since only around 5% of all records in our modeling dataset have label of 1, we're doing no better than a random guess.

What happens if we alter the cutoff for classifying a predicted probability as 1?

```python
predictions = pd.DataFrame(gs2.best_estimator_.predict_proba(features)).join(md2['best_new_music'])
predictions.columns = ['0', 'yhat', 'y']
del predictions['0']
```

```python
predictions.head()
```

```python
predictions.groupby('y').mean()
```

On average, a true "1" has a probability of 19%.

```python
from sklearn import dummy, metrics

cutoff = 0.25
predictions['ypred'] = np.where(predictions['yhat'] >= cutoff, 1, 0)
confusion = confusion_matrix(y_true=predictions['y'], y_pred=predictions['ypred'])
print confusion

accuracy = (confusion[0,0] + confusion[1,1]) / (confusion.sum() * 1.0)
print accuracy
```

```python
precision = metrics.precision_score(predictions['y'], y_pred=predictions['ypred']) 
recall = metrics.recall_score(predictions['y'], y_pred=predictions['ypred'])
print precision, recall
```

By decreasing the cutoff to 25%, we increase recall significantly, at the expense of precision.

### What about trying using a random forest classifier instead?

```python
data = md2
```

```python
features = data.iloc[:, 14:]
del features['content']
```

```python
label = data['best_new_music']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score

model = RandomForestClassifier(n_estimators = 20)
scores = cross_val_score(model, features.as_matrix(), label, scoring='roc_auc')
print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
```

Can we improve the AUC score by trying different numbers of estimators?

```python
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(model, features.as_matrix(), label, scoring='roc_auc')
print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))

auc_scores = {}
for n_trees in range(1, 100, 10):
    model = RandomForestClassifier(n_estimators = n_trees)
    scores = cross_val_score(model, features.as_matrix(), label, scoring='roc_auc')
    print('n trees: {}, CV AUC {}, Average AUC {}'.format(n_trees, scores, scores.mean()))
    auc_scores[n_trees] = scores.mean()
pd.Series(auc_scores, name='AUC Scores').plot();
```

AUC is significantly improved at around `n_estimators` = 30

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 30)

scores = cross_val_score(model, features.as_matrix(), label, scoring='roc_auc')
print('CV AUC {}, Average AUC {}'.format(scores, scores.mean()))
```

```python
model.fit(features.as_matrix(), label)
predictions_rf = pd.DataFrame(zip(model.predict(features), label))
predictions_rf
```

```python
confusion = confusion_matrix(y_true=label, y_pred=model.predict(features))
print confusion
```

```python
accuracy = (confusion[0,0] + confusion[1,1]) / (confusion.sum() * 1.0)
print accuracy
```

```python
precision = metrics.precision_score(y_true=label, y_pred=model.predict(features)) 
recall = metrics.recall_score(y_true=label, y_pred=model.predict(features))
print precision, recall
```

The random forest model has much higher recall than the logistic regression model. Its accuracy is 99.8%, so better than a random guess.

```python
from sklearn import dummy, metrics

ax = plt.subplot(111)

vals = metrics.roc_curve(label, model.predict(features))
ax.plot(vals[0], vals[1])
ax.set_ylim(0,1)
```

```python
metrics.roc_auc_score(label, model.predict(features))
```
